```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# plotting a  simple function
price = [48000, 54000, 60000, 72000, 65000]
year=[2015, 2016, 2018, 2019, 2020]
plt.plot(year, price) #(x,y) axis
plt.ylim(0,100000)
# plt.xlim(2017,2020)
```

```python
# sharma-kohli.csv plotting graph
batsman= pd.read_csv("sharma-kohli.csv")
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

### labels title

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma vs Kohli comparison')
plt.xlabel('years')
plt.ylabel('runs scored')
```

### Colors(hex) and line(width & style) and marker(size)

```python
plt.plot(batsman['index'],batsman['V Kohli'], color= 'black', markersize=4, marker= 'o', label = ' virat')
plt.plot(batsman['index'],batsman['RG Sharma'], linestyle='dashed', label= 'rohit')
# linestyle= dashed, dotted
# markers=  0, D o, markersize= 1,2,....soon
plt.title('Rohit Sharma vs Kohli comparison')

plt.xlabel('years')
plt.ylabel('runs scored')
plt.legend(loc='best')
```

### barchart
1. bivariate analysis
2. numerical vs categorial
3. use case= aggregate analysis of groups

```python
children = [10,5,20,40]
colors=['red','green','yellow', 'purple']
# Vertical barchart
plt.bar(colors,children)
```

```python
# horizontal Brchart
plt.barh(colors,children)
```

```python
#colors and label
bat=pd.read_csv('batsman_season_record.csv')
plt.bar(bat['batsman'],bat['2015'])
```

```python
plt.bar(np.arange(bat.shape[0])-0.2,bat['2015'],width=0.2)
plt.bar(np.arange(bat.shape[0]),bat['2016'],width=0.2)
plt.bar(np.arange(bat.shape[0])+0.2,bat['2017'],width=0.2)
plt.xticks(np.arange(bat.shape[0]),bat['batsman'])
```

```python
plt.bar(bat['batsman'],bat['2017'])
plt.bar(bat['batsman'],bat['2016'],botton=bat['2017'])
plt.bar(bat['batsman'],bat['2015'],bottom=(bat['2016']+bat['2016']))
```
